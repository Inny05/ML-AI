```python
import seaborn as sns
```

```python
plt.figure(figsize=(16, 8))
sns.set_theme(style = 'darkgrid')
sns.lineplot (data= suicide_dataset.groupby('target_country')['no_killed'].sum(), marker ='o')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(16, 8))
sns.set_theme(style = 'darkgrid')
sns.lineplot (data= suicide_dataset['target_country'].value_counts(), marker ='o')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(16, 8))
sns.set_theme(style = 'darkgrid')
sns.lineplot (data= suicide_dataset.groupby('target_country')['no_killed'].sum(), marker ='o')
sns.lineplot (data= suicide_dataset.groupby('target_country')['no_wounded'].sum(), marker= 'o')
plt.xticks(rotation = 90)
plt.legend(labels= ['number of killed', 'number of wounded'])
```

```python
plt.figure(figsize=(16, 8))
sns.set_theme(style = 'darkgrid')
sns.lineplot (data= suicide_dataset.groupby('target_weapon')['no_killed'].sum(), marker ='o')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(16, 8))
sns.set_theme(style = 'darkgrid')
sns.lineplot (data= suicide_dataset.groupby('target_country')['no_killed'].sum(), marker ='o')
sns.lineplot (data= suicide_dataset.groupby('target_country')['no_wounded'].sum(), marker= 'o')
sns.lineplot (data= suicide_dataset.groupby('target_country')['no_wounded'].sum(), marker= 'o')
plt.xticks(rotation = 90)
```
